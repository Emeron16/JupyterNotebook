```python
#Removes any broken images
path = Path('/Volumes/Datasets/Cars')
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
len(failed)
```

```python
path = Path('/Users/princemarcelle/Downloads/archive (1)/car_data/car_data/train')
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=32)

dls.show_batch(max_n=6)
```

```python
#Train the model
learn = vision_learner(dls, resnet18, metrics=error_rate)

# Implement EarlyStoppingCallback
# monitor='valid_loss' is common, but you could also monitor 'error_rate'
# patience=5 means it will stop if the monitored metric doesn't improve for 5 epochs
early_stop_cb = EarlyStoppingCallback(monitor='valid_loss', patience=5)

learn.fine_tune(30, cbs=[early_stop_cb])
```

```python
is_car,_,probs = learn.predict(PILImage.create('images.jpeg'))
print(f"This is a: {is_car}.")

```
