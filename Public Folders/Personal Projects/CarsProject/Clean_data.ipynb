```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('gcar_data.csv')
#drop unnamed, power_ps
df.drop(columns = ['power_ps','Unnamed: 0'], inplace = True)
print(df.shape)

df.info()

df
```

```python
df.color.fillna('unknown',inplace =True)
df.dropna(subset = ['registration_date','fuel_consumption_l_100km',\
                    'mileage_in_km','power_kw','price_in_euro','mileage_in_km'], inplace=True)
```

```python
#Remove 'km and g/Km'
df.loc[:,'fuel_consumption_g_km']= df['fuel_consumption_g_km'].str.replace('g/km','')
df.loc[:,'fuel_consumption_l_100km']=df['fuel_consumption_l_100km'].str.replace('km','')
df
```

```python
df['power_kw'] =pd.to_numeric(df['power_kw'], errors = 'coerce')
df['price_in_euro'] =pd.to_numeric(df['price_in_euro'], errors = 'coerce')
df['mileage_in_km'] =pd.to_numeric(df['mileage_in_km'], errors = 'coerce')
#convert 'fuel_consumption_g_km' and 'fuel_consumption_l_100km' to int columnns

df.info()
```

```python
#convert 'price in euro', 'power_kw', power_ps
df.loc[:,'horsepower'] = round(df['power_kw'] * 1.34102,2)
df.loc[:,'price'] = round(df['price_in_euro'] * 1.10,0)
df.loc[:,'mileage'] = round(df['mileage_in_km']* 0.621371,2)
#df.drop(columns=['power_kw','price_in_euro'], inplace = True)

```
