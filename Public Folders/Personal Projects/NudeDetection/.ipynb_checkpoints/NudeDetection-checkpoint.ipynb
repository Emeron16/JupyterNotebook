````python
from duckduckgo_search import DDGS #DuckDuckGo has changed the api so we need to update 
from fastcore.all import *
import time, json
from fastdownload import download_url
from fastai.vision.all import *
import time, json
````

````python
keywords = 'nude photos'

def search_images(keywords, max_images=200): 
    return L(DDGS().images(keywords, max_results=max_images)).itemgot('image')
````

````python
#NB: `search_images` depends on duckduckgo.com, which doesn't always return correct responses.
#    If you get a JSON error, just try running it again (it may take a couple of tries).
urls = search_images(keywords, max_images=1)
urls[0]
````

````python
path = '/Volumes/Datasets/NudeDetection/'
file_name = 'notnude.jpg'
download_url(urls[0], path + file_name, show_progress=True)

im = Image.open(path + file_name)
im.to_thumb(256,256)
````

````python
nude_keyword = 'women breast photos'
download_url(search_images(nude_keyword, max_images=1)[0], path + 'naked.jpg', show_progress=False)
Image.open(path + 'naked.jpg').to_thumb(256,256)
````
